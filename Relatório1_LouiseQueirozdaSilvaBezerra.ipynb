##Métodos

###Método bissecção

```python
def metodoBisseccao (f, a, b, i, p):
    if(f(a)*f(b) < 0):
        if (math.fabs(b - a)) < p:
            x = (a + b)/2
            return
        else:
            k = 1
            while True:
              #m = f(a)
              x = (a+b)/2
              if (f(a)*f(x) > 0):
                  a = x
              else:
                  b = x

              k += 1

              if (k-1) == i:
                  break
              elif math.fabs(f(x)) < p:
                  break
            print("Numeros de interações:", k-1)
            print("Raiz da equação:", round(x,4))
    else:
      print("Não há resultado para essa raiz.")
```

### Método da Falsa Posição

```python
def falsaPosicao(f, a, b, i, p):
  if(f(a)*f(b) < 0):
    if (math.fabs(b-a)) < p:
      x =  (a*f(b) - b*f(a)) / (f(b) - f(a))
      return
    else:
      k = 1
      while True:
        x = (a*f(b) - b*f(a)) / (f(b) - f(a))
        if math.fabs(f(x)) < p:
          break
        else:
          if (f(a) * f(x)) > 0:
            a = x
          else:
            b = x
        if math.fabs(b-a) < p:
          x = (a*f(b) - b*f(a)) / (f(b) - f(a))
          break
        k += 1

        if (k-1) == i:
          break
      print("Numeros de interações:", k-1)
      print("Raiz da equação:", round(x,4))
  else:
    print("Não há resultado para essa raiz.")
```
